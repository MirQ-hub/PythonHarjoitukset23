```python
#read csv file, append datas to lists and print first 10 rows of each list
import csv

#create empty lists for each column
Wind = []
Temp = []
Price = []

#open csv file

file = open(r"C:\Users\villemaj\OneDrive - Oulun ammattikorkeakoulu\Data Analysis\Test data\fmi_weather_and_price.csv", mode='r')#read mode

#read csv file
csv_file = csv.reader(file)

i = 0
#append columns of csv file to lists
for row in csv_file:
    if i != 0:#omit first row which is headers data
        Wind.append(float(row[1])) #save data as float (numeric instead of string)
        Temp.append(float(row[2])) 
        Price.append(float(row[3]))
    i += 1

#display first 10 rows of each list
for items in Wind[:10]:
    print(items)
for items in Temp[:10]:
    print(items)
for items in Price[:10]:
    print(items)

file.close()
```

```python
#print average of each list using inbuilt function sum() and len()

avg_wind = sum(Wind)/len(Wind) #sum of all values in list divided by number of values in list gives us average
print("Average wind speed: ", avg_wind)

avg_temp = sum(Temp)/len(Temp) #sum of all values in list divided by number of values in list gives us average
print("Average temperature: ", avg_temp)

avg_price = sum(Price)/len(Price) #sum of all values in list divided by number of values in list gives us average
print("Average electricity price: ", avg_price)
```

In statistics, a central tendency (or measure of central tendency) is a central or typical value for a probability distribution.[1]

Colloquially, measures of central tendency are often called averages. The term central tendency dates from the late 1920s.[2]

The most common measures of central tendency are the arithmetic mean, the median, and the mode. A middle tendency can be calculated for either a finite set of values or for a theoretical distribution, such as the normal distribution. Occasionally authors use central tendency to denote "the tendency of quantitative data to cluster around some central value."[2][3]

The central tendency of a distribution is typically contrasted with its dispersion or variability; dispersion and central tendency are the often characterized properties of distributions. Analysis may judge whether data has a strong or a weak central tendency based on its dispersion.
https://en.wikipedia.org/wiki/Central_tendency

```python
import numpy as np #import numpy module which contains different functions for data analysis, pip install numpy (and maybe need to restart kernel)
from scipy import stats #import stats module from scipy which contains different functions for statistical analysis, pip install scipy 

#some example data array

data = np.array([10, 15, 20, 20, 25, 30, 35, 40])

mean = np.mean(data) #calculate mean of data array
median = np.median(data) #calculate median of data array
mode = stats.mode(data) #calculate mode of data array

print("Mean: ", mean)
print("Median: ", median)
print("Mode: ", mode.mode[0]) #mode is an array, we need to access the first value of the array, different Scipy version might give different output
print("Mode: ", mode.mode) #mode is an array, we need to access the first value of the array
```

Variance and Standard Deviation:
Variance: Measures how much the data values deviate from the mean.
Standard Deviation: Square root of the variance

```python
variance = np.var(data) #calculate variance of data array
std_deviation = np.std(data) #calculate standard deviation of data array

print("Variance: ", variance)
print("Standard deviation: ", std_deviation)
```

Probability Distributions:
Python's scipy.stats module provides functions to work with various probability distributions.
Can be used to study if some value in normally distributed data falls in some area of data range.
Useful to test if some data point is outlier, for example

```python
from scipy.stats import norm #import norm module from scipy.stats which contains different functions for normal distribution, pip install scipy
import numpy as np
import seaborn as sns #import seaborn module which contains different functions for data visualization, pip install seaborn
#some example distorbution
import matplotlib.pyplot as plt #import matplotlib.pyplot 

mean = 0 #mean of distribution
std_deviation = 1 #standard deviation of distribution
value = -1 #value 

#calculate probability density function (cdf) of value

probability = 1-norm.cdf(value, loc=mean, scale=std_deviation) #norm.pdf(value, mean, standard deviation)

print("Probability: ", probability)

#sns.distplot(np.random.normal(mean, std_deviation, 1000), hist=False) #plot normal distribution with mean, standard deviation and 1000 values
sns.histplot(np.random.normal(mean, std_deviation, 1000))
#vertical line at value into distribution
plt.axvline(value, color='red')
```

Rank correlation coefficients, such as Spearman's rank correlation coefficient and Kendall's rank correlation coefficient () measure the extent to which, as one variable increases, the other variable tends to increase, without requiring that increase to be represented by a linear relationship
As we go from each pair to the next pair 
x increases, and so does y. This relationship is perfect, in the sense that an increase in 
x is always accompanied by an increase in y. This means that we have a perfect rank correlation, and both Spearman's and Kendall's correlation coefficients are 1, whereas in this example Pearson product-moment correlation coefficient is , indicating that the points are far from lying on a straight line

```python
import matplotlib.pyplot as plt
import pandas as pd

#dictionary of data
data = {'X': [1, 2, 3.5, 4.1, 5.5],
        'Y': [5, 7, 9, 11, 13]}
df = pd.DataFrame(data) #create dataframe from dictionary

correlation_matrix = df.corr('spearman') #calculate correlation matrix
print("Spearmans correlation coefficient: ", correlation_matrix)
correlation_matrix2 = df.corr('pearson') #calculate correlation matrix
print("Pearson correlation coefficient: ", correlation_matrix2)
#perfect positive correlation is 1, perfect negative correlation is -1, no correlation is 0

plt.scatter(df['X'], df['Y']) #scatter plot of data

#fit linear regression line to graph
plt.plot(np.unique(df['X']), np.poly1d(np.polyfit(df['X'], df['Y'], 1))(np.unique(df['X'])), color='red')
```

```python

import pandas as pd
import matplotlib.pyplot as plt
#Pearson correlation of signals vs phase shift
#phase shift affect to correlation but scaling or offset does not (linear operations)

Fs = 8000 #sample rate
f = 5 #frequency
sample = 8000 #number of samples
x = np.arange(sample)
y1 = np.sin(2 * np.pi * f * x / Fs)
y2 = (2*np.sin((2 * np.pi * f * x / Fs)-np.pi/2)+1) #phase shifted by pi/2-np.pi/4

fig1= plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot(x, y1, color='blue')
ax1.plot(x, y2, color='red')
ax1.set_xlabel('sample(n)')
ax1.set_ylabel('signal')
plt.show()
```
