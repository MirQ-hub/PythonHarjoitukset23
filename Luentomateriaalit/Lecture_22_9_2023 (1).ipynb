Web scraping involves extracting data from websites. Beautifulsoup-library for parsing HTML and XML documents 
pip install beautifulsoup4. There might be some restrictions for using data from websites, similarly than images etc.

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.worldometers.info/coronavirus/"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

titles = soup.find_all('h1')

for title in titles:
    print(title.text)
```

```python
url = "https://www.is.fi/"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

titles = soup.find_all('h2')

for title in titles:
    print(title.text)
```

When using open data, typical way is to use APIs (Application Programming Interfaces).
Request library can be used for making HTTP requests and APIs. There are lot of different Python libraries for APIs

```python
#https://www.tilastokeskus.fi/media/uploads/org/avoindata/kartta-aineistojen_lataus_url-osoitteen_kautta_ohje.pdf

import requests
import io

url = "https://geo.stat.fi/geoserver/wfs?service=WFS&version=2.0.0&request=GetFeature&typeName=tieliikenne:tieliikenne_2019&outputFormat=csv"
response = requests.get(url)

#read response as csv into dataframe
df = pd.read_csv(io.StringIO(response.text))
df
```
